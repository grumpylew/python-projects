### set_index and reset_index Methods
always reset first before setting a new index

```python
bond.set_index(keys = 'Film', inplace = True)
bond.head()
```

```python
bond.reset_index(inplace = True)
bond.head()
```

### .loc and .iloc 
iloc = by index, loc = by label

```python
bond = pd.read_csv('jamesbond.csv', index_col = 'Film')
bond.iloc[0]
bond.iloc[[15,20]]
bond.iloc[1:4]
bond.iloc[14,2:4]
```

```python
bond.loc['From Russia with Love']
bond.loc[['Goldfinger','Thunderball']]
```

```python
bond.loc['Moonraker','Director']
```

```python
bond.loc['Moonraker', ['Director','Box Office']]
bond.loc[['Moonraker','A View to a Kill'], ['Director','Box Office']]
```

```python
bond.loc['Moonraker', 'Director':'Budget']
```

## Set New Value for a Specific Cell

```python
bond.loc['Dr. No', 'Actor'] = 'Sir Sean Connery'
bond.loc['Dr. No']
```

```python
bond.loc['Dr. No', ['Box Office', 'Budget', 'Bond Actor Salary']] = [1,2,3]
bond.loc['Dr. No']
```

## Set Multiple Values in a DataFrame

```python
sean = bond['Actor'] == 'Sean Connery'
```

```python
bond.loc[sean, 'Actor'] == 'Sir Sean Connery'
```

### Group By

```python
fortune = pd.read_csv('fortune1000.csv', index_col = "Rank")
fortune.head()
```

```python
sectors = fortune.groupby("Sector")
sectors.head()
```

```python
sectors.get_group("Energy")
```

```python
fortune[fortune["Sector"] == "Apparel"]
```

```python
sectors["Revenue"].sum()
sectors["Profits"].max()
sectors["Profits"].min()
sectors[['Revenue','Profits']].sum()
```

```python
#group by more than 1 column
sectors = fortune.groupby(["Sector","Industry"])
```

```python
sectors.sum()
```

```python
sectors = fortune.groupby("Sector")
sectors.agg({"Revenue" : "sum",
             "Profits" : "sum",
             "Employees": "mean"})
```

```python
sectors.agg(['sum','mean'])
```
